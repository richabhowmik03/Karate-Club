```python
import networkx as nx
import matplotlib as mpl
from networkx.algorithms.community.centrality import girvan_newman
```

```python
Kc_graph = nx.karate_club_graph() # Kc_graph is a networkx graph
nx.draw(Kc_graph, with_labels=True) # Draw the graph
mpl.pyplot.show() # Show the graph
```

```python
print(nx.info(Kc_graph)) #generating the information of the graph
```

```python
pos = nx.spring_layout(Kc_graph) #positioning
nx.draw(Kc_graph, cmap = mpl.pyplot.get_cmap('inferno'), with_labels=True, pos=pos) #drawing the graph
```

```python
#Question 2
```

```python
connected_components = list(nx.connected_components(Kc_graph)) #number of connected components
print(connected_components)
```

```python
#Question 3
```

```python
communities = girvan_newman(Kc_graph) # communities is a generator of communities
print(list(communities))
```

```python
nodes = Kc_graph.nodes() #setting variable containing number of nodes
edges = Kc_graph.edges()#setting variable containing number of edges
```

```python
print(Kc_graph.nodes())#printing variable containing number of nodes
print(Kc_graph.edges())#printing variable containing number of edges
```

```python
node_groups = [] #creating node groups for segregating in coloured communities
for com in next(communities):
    node_groups.append(list(com))
print(node_groups)
```

```python
color_map = [] #visualizing the communities
for node in Kc_graph:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(Kc_graph, node_color=color_map, with_labels=True)
mpl.pyplot.show()
```

```python
import matplotlib.pyplot as plt #importing pyplot
pos = nx.spring_layout(Kc_graph)
betCent = nx.betweenness_centrality(Kc_graph, normalized=True, endpoints=True)
node_color = [20000.0 * Kc_graph.degree(v) for v in Kc_graph]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(Kc_graph, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```
